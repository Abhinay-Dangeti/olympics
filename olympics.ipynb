### Introduction

In the below project , I am analysing the olympic dataset which contains the details of people who won medals in olympics from 2000 to 2012.It also contains the category of sports they won medals and the date on which they have won medals.

### Varaibles in the Dataset

|variable name|variable explanation|
|-------------|--------------------|
|name|Name of the person who won medal in olympics|
|age|contains the age of the person|
|country|contains the country of the person who won medal|
|year|contains the year on which he won the medal|
|sports| contains the category of sports on which he won the medal|
|gold medal|contains the total medals won by each person|
|silver medal|contains the silver medals won by each person|
|brone medal|contains the bronze medal won by each person|
|total medal|contains the total medals won by each person|
|Date|contains the Date of person on which he won the medal|

#### Please list the top 10 countries with the gold medal, then top10 countries with the silver medal, then bronze medal and the total medal.
